### Load stock data

```python
# Get list of stock names
names = pd.read_csv("../data/stock_list_full.csv", header=None, names=['code'])
names = list(names['code'].str.lower())
ten_names = ['aaa','vnm','fpt','pnj','lix','ssi','vic','vcb','vnindex','vn30index']
```

```python
# Get data of each stock
def get_data(code, path):
    f = path+"/"+code+".jl"    
    df = pd.read_json(f, lines=True)
    df['transc_date'] =pd.to_datetime(df['transc_date'], dayfirst=True)
    df.index = df['transc_date']
    return df

def filter_data(df, start=None, end=None):
    df = df[(df['transc_date'] >= start) & (df['transc_date'] <= end)]
    return df
   

def daily_rate(df):
    """
    Get daily rate from a serie of adjusted/closed prices
    """
    n = df.shape[0] - 1    
    start_price = float(df[0])
    end_price = float(df[n])
    return (end_price/start_price)**(1/n) - 1

def yearly_rate(daily_rate):
    return (1+daily_rate)**(249.8) - 1
```

```python
d_rois = []
for code in names:
    df = get_data(code,path)
    data = filter_data(df, sta, end)['adjted_price']
    if data.shape[0] > 0:
        roi = daily_rate(data)
        d_rois.append([code, roi])
```

```python
df_rois = pd.DataFrame(d_rois, columns=['code','daily_roi'])
```

```python
df_rois['yearly_roi'] = df_rois['daily_roi'].apply(yearly_rate)
```

```python
daily_returns = vnm['adjted_price'].pct_change()
monthly_returns = vnm['adjted_price'].resample('M').ffill().pct_change()
yearly_returns = vnm['adjted_price'].resample('Y').ffill().pct_change()
```
